## Create Risk Metric

```python
#BusRevenue is revenue before Covid
#BusinessLossMetric = (lostExpenses-BusRevenue)/BusRevenue
#Making it into a decimal to make it the same scale as staff cancellation
Owners['BusinessLossMetric'] = (np.abs(Owners['LostExpenses_Dollars_median'] - Owners['BusiRevenue_Avg_median']))/ Owners['BusiRevenue_Avg_median']
```

```python
#Find the average of the 2 main risk components (down the col for each col)
x = Owners['BusinessLossMetric'].mean()
y = Owners['Staff_Cancellation_median'].mean()
```

```python
#Weights here are the average of each col divided by the sum of the 2 avgs
BusLossWeight = x / (x+y)
StaffCancelWeight = y / (x+y)
```

```python
#To calculate the risk, we take each value of BusLoss & staff cancellation
#multiply by the weight (how much towards risk each value contributes)
Owners['BusLossNew'] = Owners['BusinessLossMetric'] * BusLossWeight
Owners['Staff_Cancellation_new'] = Owners['Staff_Cancellation_median'] * StaffCancelWeight
```

```python
#Risk adds up the the weighted math (BusLossNew + StaffCancellationNew)
#This value is on the same scale as the components
Owners['Risk'] = Owners['Staff_Cancellation_new'] + Owners['BusLossNew']
Owners.head()
```

```python
Owners['Risk'].describe()
```

```python
#Anything above the 75th percentile will be considered risky
risk_value = np.quantile(Owners['Risk'], .75)
```

```python
Owners['Risky'] = pd.Series()
Owners.loc[Owners['Risk'] >= risk_value, 'Risky'] = 'Yes'
Owners.loc[Owners['Risk'] < risk_value, 'Risky'] = 'No'
```
