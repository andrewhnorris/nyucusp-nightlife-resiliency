```python
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
import os
import glob
```

## Read in CSVs containing yelp query results:

```python
path = #' local destination of yelp query datasets '
extension = 'csv'
os.chdir(path)
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
print('Number of listings contained in queries:', combined_csv.shape[0])
```

```python
# remove yelp index column
all_yelp = combined_csv.drop(columns=['Unnamed: 0'])
# drop duplicates
all_yelp.drop_duplicates(subset='id', inplace=True)
# print df shape, columns
print('Total number venues queried:', all_yelp.shape[0])
print('Information from query:\n', all_yelp.columns)
```

```python
# dataframe containing only information desired
yelp = all_yelp[['id', 'name', 'categories', 'rating', 'price', 'hours', 'coordinates', 'location', 'review_count', 'rev1', 'rev2', 'rev3', 'url']].copy()
yelp.reset_index(drop=True, inplace=True)   
```

```python
# zip code column
yelp['zipcode'] = (yelp.location.str.extract(r'zip_code\': \'(\d+)'))

# address column
yelp['fullAddress'] = yelp.location.str.extract('display_address\': \[([^/]*\')')
yelp['fullAddress'] = yelp['fullAddress'].str.replace("\'","")

# lat, long columns
yelp['latitude'] = yelp.coordinates.str.extract('latitude\': ([^/]*,)')
yelp['latitude'] = (yelp['latitude'].str.replace(',',''))
yelp['longitude'] = yelp.coordinates.str.extract('longitude\': ([^/]*})')
yelp['longitude'] = (yelp['longitude'].str.replace('}',''))

# all review text to single column
yelp['all_reviews'] = yelp.rev1 + " " + yelp.rev2 + " " + yelp.rev3
```

## Extracting all Yelp reviews per zipcode:

```python
# dataframe: text from all yelp reviews for each zipcode
# exported as CSV for text analysis
zip_reviews_df = yelp[['zipcode','all_reviews']]
zip_reviews_df.to_csv()
```

## Categories unpacking:

```python
#create categories index
os.chdir('/Users/andrewnorris/capstone_legaloccupancy/cleaningYelpData')
cat = pd.read_json('categories.json')

cat['aliasTitle'] = '{' + cat['alias'] + ', ' + cat['title']+ '}'
catIndex = pd.Index(list(cat['aliasTitle']))
```

```python
# get number of categories
yelp['numCats'] = yelp.categories.str.count('}, {') + 1

# data cleaning
yelp['catIndexes'] = yelp['categories'].str.replace('\'', '')
yelp['catIndexes'] = yelp['catIndexes'].str.replace('[', '')
yelp['catIndexes'] = yelp['catIndexes'].str.replace(']', '')
yelp['catIndexes'] = yelp['catIndexes'].str.replace('alias: ', '')
yelp['catIndexes'] = yelp['catIndexes'].str.replace('title: ', '')

#match categories with dictionary index
for x in yelp['catIndexes']:
    for y in catIndex: 
        if y in x:
            yelp['catIndexes'] = yelp['catIndexes'].str.replace(y, str(catIndex.get_loc(y)))

#yelp.head(10)
```

```python
# manual replacements for newamerican, tradamerican categories

yelp['newIndex'] = ""

for i in range(len(yelp['catIndexes'])):
    idxlist = list(yelp['catIndexes'][i].split(","))
    try:
        idxlist.remove(' {tradamerican')
        idxlist.append('1432') 
    except:
        pass
    try:
        idxlist.remove('{tradamerican')
        idxlist.append('1432') 
    except:
        pass
    try:
        idxlist.remove(' American (Traditional)}')
    except:
        pass
    try:
        idxlist.remove('American (Traditional)}')
    except:
        pass
    try:
        idxlist.remove(' {newamerican')
        idxlist.append('937')
    except:
        pass
    try:
        idxlist.remove('{newamerican')
        idxlist.append('937')
    except:
        pass
    try:
        idxlist.remove(' American (New)}')
    except:
        pass
    try:
        idxlist.remove('American (New)}')
    except:
        pass
    yelp['newIndex'][i] = idxlist

#yelp.head()
```

```python
# column formatting
yelp['newIndex'] = [','.join(map(str, l)) for l in yelp['newIndex']]
```

```python
# expanding list of cat indexes to separate cols
yelp = pd.concat([yelp, yelp['newIndex'].str.split(',', expand=True)], axis=1)
#yelp.tail(2)
```

```python
# rename cols
yelp = yelp.rename(columns={0: "cat0", 1: "cat1", 2: "cat2"}, errors="raise")
#yelp.head()
```

```python
# data cleaning

# value formatting
yelp.replace(to_replace=[None], value=np.nan, inplace=True)
yelp['cat0'].astype("Float32").astype("Int32")
yelp['cat1'].astype("Float32").astype("Int32")
yelp['cat2'].astype("Float32").astype("Int32")
# # category placeholder
yelp['cat0'] = yelp['cat0'].fillna("420420")
yelp['cat1'] = yelp['cat1'].fillna("420420")
yelp['cat2'] = yelp['cat2'].fillna("420420")
# value formatting
yelp['cat0'] = yelp['cat0'].str.strip()
yelp['cat1'] = yelp['cat1'].str.strip()
yelp['cat2'] = yelp['cat2'].str.strip()
```

## Category variety within each neighborhood:

```python
yelp['zipcode'] = yelp['zipcode'].fillna("0")
yelp['zipcode'] = yelp['zipcode'].astype(int)
```

```python
# map zipcode to neighborhood

#dictionary to map zips to neighborhoods
zip_dict ={ 11101:'Astoria/LIC', 11102:'Astoria/LIC', 11103:'Astoria/LIC', 11105:'Astoria/LIC', 11106:'Astoria/LIC', 11104:'LIC',
           11109:'LIC', 11120:'LIC', 11206:'Williamsburg/Bushwick/Bed-Stuy', 11211:'Williamsburg', 11249:'Williamsburg', 11207:'Bushwick',
           11221:'Bed-Stuy/Bushwick', 11237:'Bushwick', 10026:'Harlem', 10027:'Harlem', 10037:'Harlem', 10030:'Harlem', 10039:'Harlem',
           10001:'Chelsea', 10011:'Chelsea', 10018:'Chelsea', 10019:'Chelsea', 10020:'Chelsea', 10036:'Chelsea', 10002:'LES',
           10034: 'Inwood', 10040: 'Inwood', 10454: 'South Bronx', 10455: 'South Bronx', 10459: 'South Bronx', 10474: 'South Bronx', 
           11205: 'Bed-Stuy', 11216: 'Bed-Stuy', 11233: 'Bed-Stuy', 11238: 'Bed-Stuy',
           11385: 'Ridgewood', 11386: 'Ridgewood', 10109: 'Times Square', 
           11373: 'Elmhurst', 11379: 'Elmhurst', 11372: 'Jackson Heights', 
           11354: 'Flushing', 11355: 'Flushing', 11358: 'Flushing', 11222: 'Greenpoint'}
```

```python
pd.set_option('display.max_columns', 500)
# map zip to neighborhood
yelp['neighborhood'] = yelp['zipcode']
yelp = yelp.replace({'neighborhood': zip_dict})
```

```python
yelpCat = pd.crosstab(yelp.neighborhood, yelp.main_category)
# fill na's with 0
yelpCat = yelpCat.fillna(0)
# cast data type as int
yelpCat = yelpCat.astype(int)
# reset index 
yelpCat = yelpCat.reset_index()
# save df neighborhood index
neighborhood_index = yelpCat.neighborhood
# remove neighborhood col
yelpCat = yelpCat.drop(['neighborhood'],axis=1)
yelpCat.head()
```

```python
yelpCat.shape
```

```python
#new column for total num businesses/neighborhood
yelpCat['Total'] = yelpCat.sum(axis=1)
```

```python
#calculate percent of total accounted for by category
for x in range(len(yelpCat)):
    for y in range(30):
        yelpCat.iloc[x,y] = (yelpCat.iloc[x,y]/yelpCat.iloc[x,29]*100).round(2)
```

```python
yelpCat.head()
```

```python
#new table for categories
yelp_counts = yelp.groupby(['main_category']).size().reset_index(name='counts')
```

```python
#count of businesses
total = yelp_counts['counts'].sum()
```

```python
#calculate percent of total (percent distribution)
yelp_counts['nycPercentDist'] = (yelp_counts['counts']/total*100).round(2)
yelp_counts
```

Use the new column <b>yelp_counts['nycPercentDist']</b> as threshold for being counted into variety within a neighborhood.

```python
yelpCat.columns
```

```python
#new dataframe yelpCat_new, add new column for num of existing categories per neighborhood and for variety count

yelpCat_new = yelpCat.drop(['Total'],axis=1)
yelpCat_new['NumCategories'] = 0
yelpCat_new['VarietyCount'] = 0
```

```python
print(yelpCat_new.shape)
print(yelp_counts.shape) #29 cats
```

```python
yelpCat_new.head()
```

```python
#pos = count num of existing categories per neighborhood
#tally = count number of existing categories that have a percentage equal to or greater than the city distribution

for x in range(len(yelpCat_new)):
    pos = 0
    tally = 0
    for y in range(29):
        if yelpCat_new.iloc[x,y] > 0:
            pos += 1
        if ((yelpCat_new.iloc[x,y]) >= (yelp_counts.iloc[y, 2])):
            tally += 1
    yelpCat_new.iloc[x,29] = pos
    yelpCat_new.iloc[x,30] = tally
```

```python
yelpCat_new.head()
```

```python
# add neighborhood index to df
yelpCat_new['neighborhood'] = neighborhood_index
```

```python
# filter to only neighborhoods of focus
mome_neighborhoods = ['Astoria/LIC',
       'Bed-Stuy', 'Bed-Stuy/Bushwick', 'Bushwick', 'Chelsea', 'Elmhurst',
       'Flushing', 'Greenpoint', 'Harlem', 'Inwood', 'Jackson Heights',
       'LES', 'LIC', 'Ridgewood', 'South Bronx', 'Williamsburg',
       'Williamsburg/Bushwick/Bed-Stuy']
yelpCat_new = yelpCat_new[yelpCat_new['neighborhood'].isin(mome_neighborhoods)]
yelpCat_new
```

```python
yelpCat_new['VarietyPerc'] = (yelpCat_new['VarietyCount']/yelpCat_new['NumCategories']*100).round(2)
yelpCat_new
```
